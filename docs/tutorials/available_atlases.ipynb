# available atlases & resources

`yabplot` comes pre-packaged with support for several standard neuroimaging atlases. you can specify these using the `atlas` string argument in plotting functions.

```python
import yabplot as yab
yab.get_available_resources()
```

## cortical parcellations
used with `yabplot.plot_cortical(atlas='name')`.

### Desikan-Killiany (aparc)
the default FreeSurfer parcellation based on gyral and sulcal anatomy.
* **key:** `'aparc'`
* **regions:** 70 (35 per hemisphere)
* **reference:** [Desikan et al., 2006](https://doi.org/10.1016/j.neuroimage.2006.01.021)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='aparc', category='cortical')[:10])
p = yab.plot_cortical(atlas='aparc', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300), proc_vertices='sharp')
```

### Schaefer (2018)
functional parcellation based on resting-state fMRI. available in multiple granularities.
* **keys:** `'schaefer100'`, `'schaefer200'`, `'schaefer300'`, `'schaefer400'`, `'schaefer1000'`
* **reference:** [Schaefer et al., 2018](https://doi.org/10.1093/cercor/bhx179)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='schaefer400', category='cortical')[:10])
p = yab.plot_cortical(atlas='schaefer400', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

### Brainnetome (cortical)
connectivity-based parcellation using structural and functional magnetic resonance imaging.
* **key:** `'brainnetome'`
* **reference:** [Fan et al., 2016](https://doi.org/10.1093/cercor/bhw157)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='brainnetome', category='cortical')[:10])
p = yab.plot_cortical(atlas='brainnetome', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

### Automated Anatomical Labeling 3, AAL3 (cortical)
macroanatomical parcellation. the cortical subset of the full atlas.
* **key:** `'aal3'`
* **reference:** [Rolls et al., 2020](https://doi.org/10.1016/j.neuroimage.2019.116189)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='aal3', category='cortical')[:10])
p = yab.plot_cortical(atlas='aal3', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

## subcortical structures
used with `yabplot.plot_subcortical(atlas='name')`.

### FreeSurfer segmentation (aseg)
standard subcortical segmentation including Thalamus, Caudate, Putamen, Pallidum, Hippocampus, Amygdala, and Accumbens.
* **key:** `'aseg'`
* **regions:** 14 subcortical structures
* **reference:** [Fischl et al., 2002](https://doi.org/10.1016/S0896-6273(02)00569-X)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='aseg', category='subcortical')[:10])
p = yab.plot_subcortical(atlas='aseg', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

### Tian subcortical S1 (2020)
functional hierarchical parcellation of the subcortex based on resting-state fMRI. Scale 1 represents the most macroscale networks.
* **key:** `'tian2020_s1'`
* **regions:** 16 regions
* **reference:** [Tian et al., 2020](https://doi.org/10.1038/s41593-020-00711-6)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='tian2020_s1', category='subcortical')[:10])
p = yab.plot_subcortical(atlas='tian2020_s1', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

### MuSus-100
multi-modal subcortical structures atlas providing highly detailed segmentations of deep brain nuclei and thalamic subregions.
* **keys**: `'musus100'` (full 100 regions), `'musus100_dbn'` (20 deep brain nuclei), `'musus100_tha'` (thalamic subregions)
* **regions:** 20 deep brain nuclei + 80 thalamic subregions
* **reference:** [He, C. et al., 2023](https://doi.org/10.1007/s00429-022-02547-1)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='musus100_dbn', category='subcortical')[:10])
p = yab.plot_subcortical(atlas='musus100_dbn', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

### Automated Anatomical Labeling 3, AAL3 (subcortical)
macroanatomical parcellation. the subcortical subset includes deep brain structures, the cerebellum, and brainstem.
* **keys:** `'aal3'` or `'aal3_nocer'` (without cerebellum)
* **regions:** 52 subcortical regions + 26 cerebellum subregions
* **reference:** [Rolls et al., 2020](https://doi.org/10.1016/j.neuroimage.2019.116189)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='aal3', category='subcortical')[:10])
p = yab.plot_subcortical(atlas='aal3', views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300))
```

## white matter tracts
used with `yabplot.plot_tracts(atlas='name')`.

### XTRACT standard
standardised automated probabilistic tractography protocols derived from the Human Connectome Project. available in multiple downsampled resolutions for performance.
* **keys:** `'xtract_tiny'`, `'xtract_small'`, `'xtract_medium'`, `'xtract_large'`
* **regions**: 42 major white matter bundles
* **reference:** [Warrington et al., 2020](https://doi.org/10.1126/sciadv.aba8245)

```python
print('first 10 regions:', yab.get_atlas_regions(atlas='xtract_small', category='tracts')[:10])
yab.plot_tracts(atlas='xtract_small', orientation_coloring=False, views=['left_lateral', 'superior', 'left_medial'], figsize=(800, 300), bmesh='inflated')
```

## context brain meshes

standard topological brain meshes used as the background context for plotting.

**keys**: 
* `'midthickness'` (middle of the cortical ribbon) *[default]*
* `'pial'` (outer grey matter surface)
* `'white'` (inner white matter surface)
* `'swm'` (smoothed white matter)
* `'inflated'` (smoothed surface exposing sulci)
* `'very_inflated'` (spherical-like expanded surface)

```python
bmeshes = ['midthickness', 'pial', 'white', 'swm', 'inflated', 'very_inflated']
for bmesh in bmeshes:
    print(f'\n{bmesh}:')
    yab.plot_cortical(data={}, vminmax=[0, 0], views=['left_lateral', 'superior', 'left_medial'], 
                    figsize=(800, 300), bmesh=bmesh)
```
